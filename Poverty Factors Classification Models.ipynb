# About the Dataset


This dataset contains several key columns related to the Multidimensional Poverty Index (MPI) for different countries. Here's a quick overview of the columns:

* **ISO**: Country code.
* **Country**: Name of the country.
* **MPI Urban**: MPI value for urban areas.
* **Headcount Ratio Urban**: Percentage of the urban population that is * multidimensionally poor.
* **Intensity of Deprivation Urban**: Average percentage of deprivations experienced by poor people in urban areas.
* **MPI Rural**: MPI value for rural areas.
* **Headcount Ratio Rural**: Percentage of the rural population that is multidimensionally poor.
* **Intensity of Deprivation Rural**: Average percentage of deprivations experienced by poor people in rural areas.

### Scatter Plot

* **MPI Urban vs MPI Rural**
> There is a positive relationship between urban and rural MPI values, indicating that countries with higher urban MPI tend to also have higher rural MPI. However, rural MPI values are generally higher than urban ones, which implies more severe multidimensional poverty in rural areas.

* **Headcount Ratio Urban vs Headcount Ratio Rural**
> A similar trend is observed with the headcount ratio, where higher urban poverty rates are associated with higher rural poverty rates. The spread of the data points suggests a wider variation in the headcount ratio in rural areas.

* **Intensity of Deprivation Urban vs Rural**
> The intensity of deprivation between urban and rural areas shows a somewhat less distinct correlation than the MPI and headcount ratio. Nonetheless, many data points are above the line of equality, suggesting that the intensity of poverty is generally higher in rural areas.

* **Distribution of MPI Urban and Rural**
> The distribution of MPI values for urban (blue) and rural (orange) areas are overlaid. Rural areas have a broader and higher distribution, indicating higher and more varied MPI values compared to urban areas, confirming more pronounced poverty in rural areas.

* **Distribution of Headcount Ratio Urban and Rural**
> The histogram for the headcount ratio presents a clear dichotomy, with rural areas again showing higher values. This pattern signifies that a greater proportion of the rural population is affected by poverty.

* **Distribution of Intensity of Deprivation Urban and Rural**
> The intensity of deprivation shows a bimodal distribution for rural areas, suggesting the presence of two subgroups within the rural population with distinct levels of deprivation.

```python
MPIsubnationalData = pd.read_csv('/content/MPI_subnational.csv')

# Extracting National Data for Indonesia(IDN) from the national MPI dataset
indonesia_national_mpi = MPIdata[MPIdata['ISO'] == 'IDN'][['Country', 'MPI Urban', 'MPI Rural']]

# Extracting Regional Data (Provinsi) for Indonesia(IDN) from the Subnational MPI dataset
indonesia_subnational_mpi = MPIsubnationalData[MPIsubnationalData['ISO country code'] == 'IDN'][['Sub-national region', 'MPI Regional', 'Headcount Ratio Regional', 'Intensity of deprivation Regional']]

# Merging the National and Subnational Data
indonesia_national_mpi_melted = indonesia_national_mpi.melt(id_vars=['Country'], var_name='Region', value_name='MPI')
indonesia_national_mpi_melted['Region'] = indonesia_national_mpi_melted['Region'].replace({'MPI Urban': 'National (Urban)', 'MPI Rural': 'National (Rural)'})

# Concatenating and rearranging columns
indonesia_data = pd.concat([indonesia_national_mpi_melted, indonesia_subnational_mpi.rename(columns={'Sub-national region': 'Region', 'MPI Regional': 'MPI'})])[['Country', 'Region', 'Headcount Ratio Regional', 'Intensity of deprivation Regional', 'MPI']]

indonesia_data.head()
```

```python
# Replacing NaN values in 'Country' column with "Indonesia" and resetting the index
indonesia_data['Country'] = indonesia_data['Country'].fillna('Indonesia')
indonesia_data.head()
```

```python
# Separating the national and subnational data for individual visualizations
national_data = indonesia_data[indonesia_data['Region'].str.contains('National')]
subnational_data = indonesia_data[~indonesia_data['Region'].str.contains('National')]
fig, axes = plt.subplots(nrows=2, figsize=(12, 10), gridspec_kw={'height_ratios': [1, 3]})
subnationalData_sorted = subnational_data.sort_values(by='MPI', ascending= True)

# National MPI plot
sns.barplot(
    ax      = axes[0],
    x       = 'MPI',
    y       = 'Region',
    data    = national_data,
    palette = 'Greens'
)
axes[0].set_title('National MPI in Indonesia (Urban & Rural)')
axes[0].set_xlabel('MPI')

# Subnational MPI plot
sns.barplot(
    ax      = axes[1],
    x       = 'MPI',
    y       = 'Region',
    data    = subnationalData_sorted,
    palette = 'Greens')
axes[1].set_title('Subnational MPI in Indonesia (Regions)')
axes[1].set_xlabel('MPI')
axes[1].set_ylabel('Region')
plt.gca().set_facecolor('lightgray')
plt.tight_layout()
plt.show()
```

```python
# Calculate Z-score for each numeric column
z_scores = (indonesia_data.select_dtypes(include=['float64'])
             - indonesia_data.select_dtypes(include=['float64']).mean()) / indonesia_data.select_dtypes(include=['float64']).std()

# Define threshold for Z-score (e.g., 3)
threshold = 3

# Identify outliers based on Z-score
outliers = (z_scores > threshold) | (z_scores < -threshold)

# Display outliers
outlier_data = indonesia_data[outliers.any(axis=1)]
print("Outliers:")
print(outlier_data)
```

```python
from imblearn.under_sampling import RandomUnderSampler

x_ULR = df.drop('Label', axis=1)
y_ULR = df['Label']

# Split data menjadi training dan testing set
x_train_ULR, x_test_ULR, y_train_ULR, y_test_ULR = train_test_split(x_ULR, y_ULR, test_size=0.2, random_state=42)

# Check class distribution before applying SMOTE
print("Class distribution before SMOTE:")
print(y_train_ULR.value_counts())

# Gunakan RandomUnderSampler untuk melakukan undersampling pada data training
undersampler_ULR = RandomUnderSampler(random_state=42)
x_resampled_underLR, y_resampled_underLR = undersampler_ULR.fit_resample(x_train_ULR, y_train_ULR)

# Check class distribution after applying SMOTE
print("\nClass distribution after SMOTE:")
print(y_resampled_underLR.value_counts())

# Buat model klasifikasi (contoh: Logistic Regression)
model_underLR = LogisticRegression()

# Latih model pada data yang sudah diundersample
model_underLR.fit(x_resampled_underLR, y_resampled_underLR)

# Prediksi data testing
y_pred_underLR = model_underLR.predict(x_test_ULR)

# Evaluasi model
print('\nClassification Report (Undersampled)\n')
print(classification_report(y_test_ULR, y_pred_underLR, target_names=['No Poverty', 'Poverty']))
```

Coba di oversample

```python
from imblearn.over_sampling import SMOTE

x_OLR = df.drop('Label', axis=1)
y_OLR = df['Label']

# Split data menjadi training dan testing set
x_train_OLR, x_test_OLR, y_train_OLR, y_test_OLR = train_test_split(x_OLR, y_OLR, test_size=0.2, random_state=42)

# Check class distribution before applying SMOTE
print("Class distribution before SMOTE:")
print(y_train_OLR.value_counts())

# Gunakan SMOTE untuk melakukan oversampling pada data training
oversampler_LR = SMOTE(sampling_strategy='auto', random_state=42)  # You can adjust sampling_strategy
x_resampled_overLR, y_resampled_overLR = oversampler_LR.fit_resample(x_train_OLR, y_train_OLR)

# Check class distribution after applying SMOTE
print("\nClass distribution after SMOTE:")
print(y_resampled_overLR.value_counts())

# Buat model klasifikasi (contoh: Logistic Regression)
model_over_LR = LogisticRegression()

# Latih model pada data yang sudah dioversample
model_over_LR.fit(x_resampled_overLR, y_resampled_overLR)

# Prediksi data testing
y_pred_overLR = model_over_LR.predict(x_test_OLR)

# Evaluasi model
print('\nClassification Report (Oversampled)\n')
print(classification_report(y_test_OLR, y_pred_overLR, target_names=['No Poverty', 'Poverty']))
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
# Use numpy.where to apply the conditions
indonesia_data["Label"] = np.where(indonesia_data['MPI'] < 0.10, 'No Poverty', np.where(indonesia_data['MPI'] > 0.10, 'Poverty', ''))


# Create a MinMaxScaler object
scaler = MinMaxScaler()

# List of columns to scale
columns_to_scale = ['Headcount Ratio Regional', 'Intensity of deprivation Regional']

# Fit the scaler to the data and transform the data
indonesia_data[columns_to_scale] = scaler.fit_transform(indonesia_data[columns_to_scale])

indonesia_data
```

```python
# Count the Percentage Distribution for each Labesl
label_percentages = indonesia_data['Label'].value_counts(normalize=True) * 100

# Print the percentages
print(label_percentages)
```

Since the Label distribution is imbalance, im choosing Gradient Boost, since it can be more robust to class imbalance.

```python
# Model Selection
features = ['Headcount Ratio Regional',	'Intensity of deprivation Regional']
X = indonesia_data[features]
y = indonesia_data['Label']

# Splitting data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

# Define the parameter grid for Gradient Boosting
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'min_samples_split': [2, 4],
    'min_samples_leaf': [1, 2],
    'max_features': [None, 'sqrt', 'log2']
}

# Create Gradient Boosting classifier
gbc = GradientBoostingClassifier(random_state=42)

# Instantiate the GridSearchCV object
gbc_GScv = GridSearchCV(estimator=gbc, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2, scoring='accuracy')

# Fit the GridSearchCV object to the data
gbc_GScv.fit(X_train, y_train)

# Best parameters
print("Best parameters found: ", gbc_GScv.best_params_)
```
